```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("inter_updated.xlsx", header=None)
```

```python
score_list = []
for i in range(20):
    print("i: ", i)
    for j in range(i, 19):
        print("j: ",j)
        score = df.iloc[i,j]
        if type(score) == str:
            score_list.append(score.replace(' ', ''))
```

```python
score_dict = {'+':0.33, '++': 0.66, '+++': 1, '+++(+)': (1+1.33)/2,'+/-':0, '-':-0.33, '0/+':(0.33/2) }
```

```python
inter_subset = []
inter_effect = []

for i in range(20):
    for j in range(i, 19):
        score = df.iloc[i,j]
        if type(score) == str:
            score = score.replace(' ', '')
            inter_subset.append((i,j))
            try:
                inter_effect.append(score_dict[score])
            except:
                pass
```
